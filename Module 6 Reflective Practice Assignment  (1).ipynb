**GRADED TASK 2:**  Use suitable R functions to do the following tasks:
- Find the 38th percentile value of the distribution for the scenario you described.
- Simulate 500 values from the distribution for the scenario you described. Plot the estimated probability density function and the histogram of the random variable.

```r
# YOUR CODE HERE
# Parameters
mean <- 30
sd <- 5

# Find the 38th percentile
percentile_38 <- qnorm(0.38, mean, sd)
percentile_38

# Simulate 500 values
set.seed(123)  # For reproducibility
simulated_values <- rnorm(500, mean, sd)

# Plot the histogram
hist(simulated_values, breaks = 30, probability = TRUE, 
     main = "Histogram and Estimated PDF of Daily Coffee Consumption",
     xlab = "Coffee Consumption (ounces)", col = "lightblue")

# Overlay the estimated PDF
lines(density(simulated_values), col = "red", lwd = 2)

# Add a legend
legend("topright", legend = c("Histogram", "Estimated PDF"), 
       col = c("lightblue", "red"), lwd = 2, bty = "n")

```
